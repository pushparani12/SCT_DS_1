```python
# Step 1: Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the CSV file into a DataFrame
df = pd.read_csv('/content/archive (2).zip')

# Step 3: Inspect the first few rows of the dataset to confirm itâ€™s loaded correctly
print(df.head())

# Step 4: Group the data by 'race/ethnicity' and 'parental level of education'
race_edu_counts = df.groupby(['race/ethnicity', 'parental level of education']).size().unstack()

# Step 5: Create the Stacked Bar Chart for Race/Ethnicity vs Parental Level of Education
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the stacked bar chart
race_edu_counts.plot(kind='bar', stacked=True, ax=ax)

# Step 6: Set the labels, title, and legend
ax.set_xlabel('Race/Ethnicity')
ax.set_ylabel('Count')
ax.set_title('Distribution of Parental Level of Education by Race/Ethnicity')

# Move the legend outside the plot for better readability
ax.legend(title='Parental Level of Education', bbox_to_anchor=(1.05, 1), loc='upper left')

# Step 7: Adjust the layout for better spacing and display the chart
plt.tight_layout()
plt.show()
```
